# Embedding

````python
import torch
import torch.nn as nn
````

## Step-1 : build vocabulary

````python
vocab = {
    "<PAD>": 0,
    "<UNK>": 1,
    "I": 2,
    "love": 3,
    "deep": 4,
    "learning": 5
}
````

## Step-2 : Sentence to Token Ids

````python
token_ids = torch.tensor([2, 3, 4, 5])
print("Token IDs:")
print(token_ids)
print("Shape:", token_ids.shape)
````

## Step-3 : Embedding Layer

````python
embedding=nn.Embedding(
    num_embeddings=len(vocab),
    embedding_dim=4
)
print("\nEmbedding Weight Matrix:")
print(embedding.weight)
print("Shape:", embedding.weight.shape)
````

## Step-4 : Embedded Lookup

````python
embedded=embedding(token_ids)
print("\nEmbedded Lookup:")
print(embedded)
print("Shape:", embedded.shape)
````
