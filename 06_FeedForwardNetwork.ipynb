# Feed Forward Network

```python
import torch
import torch.nn as nn
```

```python
class FeedForward(nn.Module):
  def __init__(self,d_model,hidden_dim):
    super().__init__()

    self.fc1=nn.Linear(d_model,hidden_dim)
    self.relu=nn.ReLU()
    self.fc2=nn.Linear(
        hidden_dim,d_model
    )

  def forward(self,x):
    out=self.fc1(x)
    out=self.relu(out)
    out=self.fc2(out)
    return out
```

```python
batch=32
seq_lem=50
embed_dim=512
hidden_dim=2048
x=torch.randn(
    batch,seq_lem,embed_dim
)
x.shape
```

```python
ffn=FeedForward(embed_dim,hidden_dim)
output=ffn(x)
print("Input Shape: ",x.shape)
print("output Shape: ",output.shape)
```

| Multi-Head Attention               | Feed Forward Network                           |
| ---------------------------------- | ---------------------------------------------- |
| Word ↔ Word Interaction            | প্রতিটি Word আলাদাভাবে Process হয়             |
| Context শেখে                       | Feature Transform করে                          |
| Relationship শেখে                  | Representation সমৃদ্ধ করে                      |
| Token-গুলোর মধ্যে তথ্য আদান-প্রদান | একই MLP প্রতিটি Token-এর উপর আলাদাভাবে প্রয়োগ |
